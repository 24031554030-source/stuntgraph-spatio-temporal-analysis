```python
def clean_pengangguran(df, tahun):

    df = df.iloc[3:].copy()

    df.columns = [
        'Provinsi',
        'Februari',
        'Agustus',
        'Tahunan'
    ]

    df = df[['Provinsi', 'Agustus']]

    df.columns = [
        'Provinsi',
        'Pengangguran'
    ]

    df['Tahun'] = tahun

    return df
```

```python
pengangguran_2021_clean = clean_pengangguran(
    pengangguran_2021,
    2021
)

pengangguran_2022_clean = clean_pengangguran(
    pengangguran_2022,
    2022
)

pengangguran_2023_clean = clean_pengangguran(
    pengangguran_2023,
    2023
)

pengangguran_2024_clean = clean_pengangguran(
    pengangguran_2024,
    2024
)
```

```python
pengangguran = pd.concat([
    pengangguran_2021_clean,
    pengangguran_2022_clean,
    pengangguran_2023_clean,
    pengangguran_2024_clean
], ignore_index=True)
```

```python
def clean_sanitasi(df, tahun):
    df = df.iloc[2:].copy()
    df.columns = ['Provinsi', 'Sanitasi']
    df = df.dropna()
    df['Tahun'] = tahun
    return df
```

```python
def clean_air(df, tahun):

    df = df.iloc[3:].copy()

    df.columns = [
        'Provinsi',
        'Perkotaan',
        'Perdesaan',
        'Total'
    ]

    df = df[['Provinsi', 'Total']]

    df.columns = [
        'Provinsi',
        'Air_Bersih'
    ]

    df = df.dropna()

    df['Tahun'] = tahun

    return df
```

```python
air_2021_clean = clean_air(air_2021, 2021)
air_2022_clean = clean_air(air_2022, 2022)
air_2023_clean = clean_air(air_2023, 2023)
air_2024_clean = clean_air(air_2024, 2024)
```

```python
air = pd.concat([
    air_2021_clean,
    air_2022_clean,
    air_2023_clean,
    air_2024_clean
], ignore_index=True)
```

```python
def samakan_provinsi(df):
    df = df.copy()

    df['Provinsi'] = (
        df['Provinsi']
        .str.upper()
        .str.strip()
        .replace({
            'KEP. BANGKA BELITUNG': 'KEPULAUAN BANGKA BELITUNG',
            'KEP. RIAU': 'KEPULAUAN RIAU'
        })
    )
    return df

stunting = samakan_provinsi(stunting)
pendidikan = samakan_provinsi(pendidikan)
kemiskinan = samakan_provinsi(kemiskinan)
pdrb = samakan_provinsi(pdrb)
pengangguran = samakan_provinsi(pengangguran)
sanitasi = samakan_provinsi(sanitasi)
air = samakan_provinsi(air)
```

```python
df_final = stunting.merge(
    pendidikan,
    on=['Provinsi', 'Tahun'],
    how='inner'
)

df_final = df_final.merge(
    kemiskinan,
    on=['Provinsi', 'Tahun'],
    how='inner'
)

df_final = df_final.merge(
    pdrb,
    on=['Provinsi', 'Tahun'],
    how='inner'
)

df_final = df_final.merge(
    pengangguran,
    on=['Provinsi', 'Tahun'],
    how='inner'
)

df_final = df_final.merge(
    sanitasi,
    on=['Provinsi', 'Tahun'],
    how='inner'
)

df_final = df_final.merge(
    air,
    on=['Provinsi', 'Tahun'],
    how='inner'
)
```

## **9. Model Evaluation**

```python
from sklearn.metrics import silhouette_score, silhouette_samples

feature_cols = ['Kemiskinan', 'Pengangguran', 'Sanitasi', 'Air_Bersih', 'Pendidikan', 'PDRB']
df_prov = df_final.groupby('Provinsi')[feature_cols + ['Stunting']].mean().reset_index()

scaler = StandardScaler()
X = scaler.fit_transform(df_prov[feature_cols])

# Re-fit K-Means
kmeans = KMeans(n_clusters=k_optimal, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)
df_prov['Cluster'] = labels

cluster_stunting_mean = df_prov.groupby('Cluster')['Stunting'].mean().sort_values()
label_map = {}
risk_levels = ['Low Risk', 'Medium Risk', 'High Risk']
for i, (cluster_id, _) in enumerate(cluster_stunting_mean.items()):
    label_map[cluster_id] = risk_levels[i]

df_prov['Risiko'] = df_prov['Cluster'].map(label_map)
CLUSTER_COLORS = {
    'Low Risk': '#2ecc71',
    'Medium Risk': '#f39c12',
    'High Risk': '#e74c3c'
}
```

```python
# Calculate the Silhouette Score
sil_score_overall = silhouette_score(X, labels)
print(f"Silhouette Score (K={k_optimal}): {sil_score_overall:.4f}")
print("Interpretation:")
if sil_score_overall >= 0.7:
    print("STRONG clustering structure (≥ 0.7)")
elif sil_score_overall >= 0.5:
    print("MODERATE clustering structure (0.5–0.7)")
elif sil_score_overall >= 0.25:
    print("WEAK clustering structure (0.25–0.5)")
else:
    print("No significant structure (< 0.25)")
```

```python
# Silhouette Score by cluster
sample_sil = silhouette_samples(X, labels)
df_prov['Silhouette'] = sample_sil

print("Silhouette Score by Cluster:")
for cluster_id in sorted(df_prov['Cluster'].unique()):
    mask = df_prov['Cluster'] == cluster_id
    mean_sil = df_prov.loc[mask, 'Silhouette'].mean()
    risk = label_map[cluster_id]
    n = mask.sum()
    print(f"Cluster {cluster_id} ({risk}) | n={n:2d} | Silhouette = {mean_sil:.4f}")
```

```python
# Comparison of Silhouettes of Various K
K_range = range(2, 8)
sil_scores = []
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    lbl = km.fit_predict(X)
    sil_scores.append(silhouette_score(X, lbl))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Model Evaluation: Silhouette Analysis', fontsize=14, fontweight='bold')

# Plot 1: Silhouette score and K
ax1 = axes[0]
ax1.plot(list(K_range), sil_scores, marker='o', color='steelblue', linewidth=2, markersize=8)
ax1.axvline(x=k_optimal, color='red', linestyle='--', linewidth=1.5, label=f'K={k_optimal} choosen')
ax1.scatter([k_optimal], [sil_scores[k_optimal - 2]], color='red', zorder=5, s=120)
ax1.set_title('Silhouette Score by K')
ax1.set_xlabel('Number of Clusters (K)')
ax1.set_ylabel('Silhouette Score')
ax1.legend()
ax1.grid(True, alpha=0.4)

# Plot 2: Silhouette plot per sample
ax2 = axes[1]
y_lower = 10
for cluster_id in sorted(np.unique(labels)):
    risk = label_map[cluster_id]
    color = CLUSTER_COLORS[risk]
    sil_vals = np.sort(sample_sil[labels == cluster_id])
    size = len(sil_vals)
    y_upper = y_lower + size
    ax2.fill_betweenx(np.arange(y_lower, y_upper), 0, sil_vals,
                      facecolor=color, edgecolor='none', alpha=0.8)
    ax2.text(-0.05, y_lower + 0.5 * size, f'C{cluster_id}\n({risk})', fontsize=7)
    y_lower = y_upper + 10

ax2.axvline(x=sil_score_overall, color='black', linestyle='--', linewidth=1.5,
            label=f'Avg = {sil_score_overall:.3f}')
ax2.set_title(f'Silhouette Plot by Province (K={k_optimal})')
ax2.set_xlabel('Silhouette Coefficient')
ax2.set_ylabel('Province')
ax2.legend(loc='upper right')
ax2.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
